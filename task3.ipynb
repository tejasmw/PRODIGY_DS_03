# Task 3 :  Decision Tree Classifier 

Importing libraries

We will use the 'Breast Cancer' Dataset from Scikit-learn Library

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer 
```

```python
cancer = load_breast_cancer()
```

Training the model 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target,stratify = cancer.target,random_state = 42)
tree = DecisionTreeClassifier(random_state = 4)
```

```python
tree.fit(X_train,y_train)
predicted = tree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(predicted,y_test)
print('Accuracy Score : ',accuracy)
```

```python
print("Accuracy on training set : ", tree.score(X_train, y_train))
print("Accuracy on testing set : ",tree.score(X_test,y_test))
```

Visualising the important features, which determine the classification 

```python
#visualizing the feature importances :
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import numpy as np

def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.show()

plot_feature_importances_cancer(tree)
```

Visualising the decision tree  

```python
#visualising the tree
plt.figure(figsize=(12,10))
from sklearn.tree import plot_tree

plot_tree(tree, feature_names = cancer.feature_names.tolist(), class_names = cancer.target_names.tolist(), filled = True )
plt.show()
```
